### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda_1 = tienda['Precio'].sum()
ingreso_tienda_2 = tienda2['Precio'].sum()
ingreso_tienda_3 = tienda3['Precio'].sum()
ingreso_tienda_4 = tienda4['Precio'].sum()
```

```python
print('Ingreso total tienda 1: ', ingreso_tienda_1)
print('Ingreso total tienda 2: ', ingreso_tienda_2)
print('Ingreso total tienda 3: ', ingreso_tienda_3)
print('Ingreso total tienda 4: ', ingreso_tienda_4)
```

# 2. Ventas por categoría

```python
ventas_tienda_1 = tienda.groupby('Categoría del Producto')['Precio'].count().reset_index(name='Cantidad Vendidas')
ventas_tienda_2 = tienda2.groupby('Categoría del Producto')['Precio'].count().reset_index(name='Cantidad Vendidas')
ventas_tienda_3 = tienda3.groupby('Categoría del Producto')['Precio'].count().reset_index(name='Cantidad Vendidas')
ventas_tienda_4 = tienda4.groupby('Categoría del Producto')['Precio'].count().reset_index(name='Cantidad Vendidas')
```

```python
print('Ventas por categoría tienda 1: ')
print(ventas_tienda_1)
print('Ventas por categoría tienda 2: ')
print(ventas_tienda_2)
print('Ventas por categoría tienda 3: ')
print(ventas_tienda_3)
print('Ventas por categoría tienda 4: ')
print(ventas_tienda_4)
```

# 3. Calificación promedio de la tienda


```python
promedio_tienda_1 = tienda['Calificación'].mean()
promedio_tienda_2 = tienda2['Calificación'].mean()
promedio_tienda_3 = tienda3['Calificación'].mean()
promedio_tienda_4 = tienda4['Calificación'].mean()
```

```python
print('Calificación promedio tienda 1: ',round(promedio_tienda_1, 2))
print('Calificación promedio tienda 2: ',round(promedio_tienda_2, 2))
print('Calificación promedio tienda 3: ',round(promedio_tienda_3, 2))
print('Calificación promedio tienda 4: ',round(promedio_tienda_4, 2))
```

# 4. Productos más y menos vendidos

```python
producto_tienda_1 = tienda.groupby('Producto')['Precio'].count().reset_index(name='Cantidad Vendidas')
producto_tienda_1 = producto_tienda_1.sort_values(by='Cantidad Vendidas', ascending=False)
producto_tienda_2 = tienda2.groupby('Producto')['Precio'].count().reset_index(name='Cantidad Vendidas')
producto_tienda_2 = producto_tienda_2.sort_values(by='Cantidad Vendidas', ascending=False)
producto_tienda_3 = tienda3.groupby('Producto')['Precio'].count().reset_index(name='Cantidad Vendidas')
producto_tienda_3 = producto_tienda_3.sort_values(by='Cantidad Vendidas', ascending=False)
producto_tienda_4 = tienda4.groupby('Producto')['Precio'].count().reset_index(name='Cantidad Vendidas')
producto_tienda_4 = producto_tienda_4.sort_values(by='Cantidad Vendidas', ascending=False)
```

```python
print('Tienda 1 - Más vendido:\n', producto_tienda_1.head(1),
      '\nMenos vendido:\n', producto_tienda_1.tail(1), '\n')
print('Tienda 2 - Más vendido:\n', producto_tienda_2.head(1),
      '\nMenos vendido:\n', producto_tienda_2.tail(1), '\n')
print('Tienda 3 - Más vendido:\n', producto_tienda_3.head(1),
      '\nMenos vendido:\n', producto_tienda_3.tail(1), '\n')
print('Tienda 4 - Más vendido:\n', producto_tienda_4.head(1),
      '\nMenos vendido:\n', producto_tienda_4.tail(1), '\n')
```

```python
import pandas as pd

def obtener_resumen_tienda(df_producto, nombre_tienda, resumen_dict):

    if df_producto.empty:
        resumen_dict[nombre_tienda] = {
            'Más Vendido': None,
            'Ventas Más Vendido': 0,
            'Menos Vendido': None,
            'Ventas Menos Vendido': 0,
        }
        return resumen_dict

    resumen_dict[nombre_tienda] = {
        'Más Vendido': df_producto.head(1)['Producto'].values[0],
        'Ventas Más Vendido': int(df_producto.head(1)['Cantidad Vendidas'].values[0]),
        'Menos Vendido': df_producto.tail(1)['Producto'].values[0],
        'Ventas Menos Vendido': int(df_producto.tail(1)['Cantidad Vendidas'].values[0]),
    }
    return resumen_dict

resumen_ventas = {}


productos = [producto_tienda_1, producto_tienda_2, producto_tienda_3, producto_tienda_4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for producto_df, nombre in zip(productos, nombres):
    resumen_ventas = obtener_resumen_tienda(producto_df, nombre, resumen_ventas)

df_resumen_funcional = pd.DataFrame(resumen_ventas).T
print(df_resumen_funcional)
```

# 5. Envío promedio por tienda

# 6. Generando Grafico


#### 1. Gráfico de barras — Ingreso total por tienda

```python
ingresos = [ingreso_tienda_1, ingreso_tienda_2, ingreso_tienda_3, ingreso_tienda_4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(8, 5))
plt.bar(tiendas, ingresos, color = ['#5DADE2','#58D68D','#F5B041','#AF7AC5'])
plt.title('Ingresos Totales por Tienda')
plt.ylabel('Ingresos')
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```

#### 2. Gráfico circular (pastel): Distribución de categorías - Tienda 1

```python
ventas_categoria_tienda_1 = tienda.groupby('Categoría del Producto')['Precio'].count()
ventas_ordenadas = ventas_categoria_tienda_1.sort_values(ascending=False)

colores_elegidos = ['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3', '#a6d854', '#ffd92f', '#e5c494', '#b3b3b3']
plt.figure(figsize=(8, 8))

plt.pie(
    ventas_ordenadas,
    labels=ventas_ordenadas.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colores_elegidos,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.5}
)
plt.title('Distribución de Categorías de Producto en Tienda 1', fontsize=14)
plt.axis('equal')
plt.show()
```

#### 3.Gráfico de líneas: Costo de envío promedio por tienda

```python
envios_promedio = [envio_prom_tienda_1, envio_prom_tienda_2, envio_prom_tienda_3, envio_prom_tienda_4]
plt.figure(figsize=(8, 5))
plt.plot(tiendas, envios_promedio, marker='o', linestyle='-', color ='#E67E22')
plt.title('Costo de Envío Promedio por Tienda')
plt.ylabel('Costo de Envío Promedio')
plt.grid(True,linestyle='--', alpha=0.6)
plt.show()
```

4️⃣ Productos Más y Menos Vendidos

Se identificaron los productos con mayor y menor demanda en cada tienda:

```python
import pandas as pd

def obtener_resumen_tienda(df_producto, nombre_tienda, resumen_dict):
    """Extrae datos de producto y agrega al diccionario de resumen."""
    resumen_dict[nombre_tienda] = {
        'Más Vendido': df_producto.head(1)['Producto'].values[0],
        'Ventas Más Vendido': df_producto.head(1)['Cantidad Vendidas'].values[0],
        'Menos Vendido': df_producto.tail(1)['Producto'].values[0],
        'Ventas Menos Vendido': df_producto.tail(1)['Cantidad Vendidas'].values[0],
    }
    return resumen_dict

resumen_ventas = {}


productos = [producto_tienda_1, producto_tienda_2, producto_tienda_3, producto_tienda_4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for producto_df, nombre in zip(productos, nombres):
    resumen_ventas = obtener_resumen_tienda(producto_df, nombre, resumen_ventas)

df_resumen_funcional = pd.DataFrame(resumen_ventas).T
print(df_resumen_funcional)
```

Observaciones:

Productos de alta demanda aseguran ventas constantes, mientras que los menos vendidos pueden necesitar promoción o reconsideración de inventario.

Tienda 3 destaca por tener productos muy demandados.
